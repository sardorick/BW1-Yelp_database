```python
names = []
ser = pd.read_csv('restaurant_soup').drop(columns='Unnamed: 0').squeeze()
for row in ser:
    soup = BeautifulSoup(row, 'html.parser')
    result = soup.select('.css-1egxyvc .css-1m051bw')
    for tag in result:
        names.append(tag.text)

pd.Series(names)
```

```python
price_range = [] 

for row in ser:
    soup = BeautifulSoup(row, 'html.parser')
    result = soup.select('.priceRange__09f24__mmOuH')
    for tag in result:
        price_range.append(tag.text)
    
pd.Series(price_range)
```

```python
review_count = []

for row in ser:
    soup = BeautifulSoup(row, 'html.parser')
    result = soup.select('.reviewCount__09f24__tnBk4')
    for tag in result:
        review_count.append(tag.text)
    
pd.Series(review_count)
```

```python
category = []

for row in ser:
    soup = BeautifulSoup(row, 'html.parser')
    result = soup.select('.css-dd1rsv')
    for tag in result:
        category.append(tag.text)
    
pd.Series(category)
```

```python
all = []
for row in ser:
    soup = BeautifulSoup(row, 'html.parser')
    result = soup.select('.reviewCount__09f24__tnBk4 , .css-1s7bx9e , .css-dzq7l1 .css-chan6m , .css-dd1rsv , .css-1egxyvc')
    
    for tag in result:
        raw_data = ''
        raw_data += tag.text + ':'
        all.append(tag.text.split('\xa0')[0].split("'"))


print(pd.DataFrame(all))
```

```python
neighbourhood = []
for row in ser:
    soup = BeautifulSoup(row, 'html.parser')
    result = soup.select('.css-dzq7l1 .css-chan6m')
    for tag in result:
        neighbourhood.append(tag.text)
    
pd.Series(neighbourhood)
```

```python
rest_df = pd.DataFrame(zip(names, neighbourhood, price_range, review_count, category), columns=['name', 'neighbourhood' , 'price_range', 'review_count', 'category'])
rest_df.to_csv('restaurant_details.csv')
rest_df.shape
rest_df
```
